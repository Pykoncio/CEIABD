## Consulta 0. Libros con más reseñas positivas que negativas

````python
%%writefile proceso_libros.pig

-- Cargar datos de libros desde el csv
bookslarge = LOAD 'temp/books-large.csv' using PigStorage(';') AS
(id:int, book_id:int, best_book_id:int, work_id:int, books_count:int, isbn:chararray, isbn13:long, authors:chararray,
original_publication_year:int, original_title:chararray, title:chararray, language_code:chararray, average_rating:float,
ratings_count:int, work_ratings_count:int, work_text_reviews_count:int,
ratings_1:int, ratings_2:int, ratings_3:int, ratings_4:int, ratings_5:int,
image_url:chararray, small_image_url:chararray);

-- separar entradas por cada autor
allbooks = FOREACH bookslarge generate isbn13 as isbn13, title as title,
average_rating as average_rating, language_code as language_code,
original_publication_year as year, ratings_count as ratings_count, ratings_1,
ratings_2, ratings_3, ratings_4, ratings_5, FLATTEN(TOKENIZE(authors, ',')) as author;

-- Libros con más reseñas positivas (5) que negativas (1/2*)
goodBooks = FILTER allbooks BY (ratings_1 + ratings_2 <= ratings_5);
goodBooks = FOREACH goodBooks generate title, author, year, ratings_1, ratings_2, ratings_5, average_rating;
goodBooksByYear = GROUP goodBooks BY year;

-- almacenar (directorio no existente)
STORE goodBooksByYear INTO 'results/' USING PigStorage (',');
````

## Consulta 1. Autores con mayor puntuación media

1. En primer lugar agrupamos los libros por autor.

>`allbooksByAuthor = group allbooks by author;`

2. Para cada autor calculamos su puntuación media y su número de libros.
Si quisiéramos medir la calidad del autor por su mejor libro, usaríamos MAX en vez de AVG

>`authorsAverage = foreach allbooksByAuthor generate group as author, AVG(allbooks.average_rating), COUNT(allbooks.average_rating);`

3. Filtramos para obtener al mejor autor.

>`bestauthor = LIMIT (order authorsAverage by $1 desc) 5;`

Si quisiéramos que el escritor haya escrito varios libros, añadimos esta línea:

>`bestauthor = filter authorsAverage by $2 > 1;`

4. Almacenamiento o mostrar el resultado

>`dump bestauthor;`

## Consulta 2. Sagas con máyor puntuación media

1. Detectar Sagas (títulos con paréntesis y una coma)

>`series = filter allbooks by (title MATCHES '(.*\\(.*,.*)');`

2. cambiar titulos a nombres de sagas, paso 1: split por posición en la saga '#'

>`seriesNamed = foreach series generate (chararray)STRSPLIT(title, ', #', 2).$0 as seriesName, average_rating; `

3. Cambiar titulos a nombres de sagas, paso 2: split por nombre de saga '('

>`seriesNamed2 = foreach seriesNamed generate (chararray)STRSPLIT(seriesName, '\\(', 2).$1 as seriesName, average_rating;`

4. Agrupación

>`seriesByName = group seriesNamed2 by seriesName;`
5. cálculo de nota media por cada serie

>`bestSeries = foreach seriesByName generate group as seriesName, AVG(seriesNamed2.average_rating) as avgRating;`

6. Mostramos las 5 sagas con mejor nota media

> `bestSeries = LIMIT(order bestSeries by avgRating desc) 5;`

> `dump bestSeries;`
