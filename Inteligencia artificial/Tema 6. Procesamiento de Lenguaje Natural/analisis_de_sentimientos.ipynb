# Análisis de sentimientos

Usaremos los pipelines de transformers:

<https://huggingface.co/docs/transformers/main_classes/pipelines>

```python
from transformers import pipeline
```

```python
# Crea una instancia del clasificador de texto
classifier = pipeline("sentiment-analysis")
```

```python
# Texto que queremos analizar
text = "I love coding in Python"
```

```python
# Ejecución del clasificador
result = classifier(text)
```

```python
result
```

```python
# Texto negativo que queremos analizar
text = "I hate myself and I want to die."
result = classifier(text)
result
```

```python
# Texto neutro que queremos analizar
text = "The computer is on."
result = classifier(text)
result
```
