# Python Regular Expression - Exercises, Practice, Solution

```python
import re

text = "Antes de nada, 0las 00actas 000de evalu_ación s0on ab abb abbb abbbb producto_ 01de u01na cultura_ heteropatriarcal y transfóbica. La lucha feminista contribuye a fortalecer los estereotipos de género. Al objeto de unificar criterios para adoptar las decisiones que se ASfdsjfj47831_ 1 1ASconsideran12_ 2relevantes891A_12, una forma 1infaliblez de1 a1segurar z la igualzdad podrá solicitar la resolución de su contrato sin coste adicional alguno."
```

```python
# 1

x = re.search("[^a-zA-Z0-9]", text)

if x:
  print("Se ha encontrado el patron")
else:
  print("No se ha encontrado el patron")
```

```python
# 2

x = re.findall("ab*", text)

print(x)

if x:
  print("Se ha encontrado el patron")
else:
  print("No se ha encontrado el patron")
```

```python
# 3

x = re.findall("ab+", text)

print(x)

if x:
  print("Se ha encontrado el patron")
else:
  print("No se ha encontrado el patron")
```

```python
# 4

x = re.findall("ab?", text)

print(x)

if x:
  print("Se ha encontrado el patron")
else:
  print("No se ha encontrado el patron")
```

```python
# 5

x = re.findall("ab{3}", text)

print(x)

if x:
  print("Se ha encontrado el patron")
else:
  print("No se ha encontrado el patron")
```

```python
# 6

x = re.findall("ab{2,3}", text)

print(x)

if x:
  print("Se ha encontrado el patron")
else:
  print("No se ha encontrado el patron")
```

```python
# 7

x = re.findall(r"[a-z]_+[a-z]+", text)

print(x)

if x:
  print("Se ha encontrado el patron")
else:
  print("No se ha encontrado el patron")
```

```python
# 8

x = re.findall(r"[A-Z][a-z]+", text)

print(x)

if x:
  print("Se ha encontrado el patron")
else:
  print("No se ha encontrado el patron")
```

```python
# 9

x = re.findall(r"a[a-zA-Z]+b", text) # ^$ no nos vale en nuestro texto ya que busca solo el principio y el fin de este

print(x)

if x:
  print("Se ha encontrado el patron")
else:
  print("No se ha encontrado el patron")
```

```python
# 10

x = re.findall("^\w+", text)

print(x)

if x:
  print("Se ha encontrado el patron")
else:
  print("No se ha encontrado el patron")
```

```python
# 11

x = re.findall("\w+\S$", text)

print(x)

if x:
  print("Se ha encontrado el patron")
else:
  print("No se ha encontrado el patron")
```

```python
# 12

x = re.findall("\w*z\w*", text)

print(x)

if x:
  print("Se ha encontrado el patron")
else:
  print("No se ha encontrado el patron")
```

```python
# 13

x = re.search("\Bz\B", text)

print(x)

if x:
  print("Se ha encontrado el patron")
else:
  print("No se ha encontrado el patron")
```

```python
# 14

x = re.findall("^[a-zA-Z0-9_]*$", "Aa_8")

print(x)

if x:
  print("Se ha encontrado el patron")
else:
  print("No se ha encontrado el patron")
```

```python
# 15

x = re.findall("\B1\w+", text)

print(x)

if x:
  print("Se ha encontrado el patron")
else:
  print("No se ha encontrado el patron")
```

```python
# 16

x = re.sub("\.[0]*", ".", "216.08.094.196")

print(x)

if x:
  print("Se ha encontrado el patron")
else:
  print("No se ha encontrado el patron")
```
