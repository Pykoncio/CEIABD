# Ejemplos de *lookbehind* y *lookahead*

```python
import re
```

## Lookbehind

La notación es `(?<=...)`

Verifica que haya un patrón antes de la coincidencia, sin incluirlo en el resultado.

```python
texto = "En USA, el precio de un Ford Mustang es de $31920 y un Tesla Model 3 cuesta $29990."
```

```python
precios = re.findall(r"\$\d+", texto)
print(precios)
```

```python
precios = re.findall(r"(?<=\$)\d+", texto) # No incluye el dolar en el resultado
print(precios)
```

## Lookahead


La notación es `(?=...)`

Verifica que haya un patrón después de la coincidencia, sin incluirlo en el resultado.

```python
texto2 = "Málaga, Alhaurín de la Torre, Marbella y Benalmádena."
```

```python
# Buscar palabras seguidas de una coma (se incluye la coma en el resultado)

palabras = re.findall(r"\w+,", texto2)
print(palabras)
```

```python
# (no se incluye la coma en el resultado)

palabras = re.findall(r"\w+(?=,)", texto2)
print(palabras)
```
