## Sentencia Condicional

```python
x = 89

if x == 88:
  print("X vale 88")
else:
  print("X no vale 88")
```

```python
edad = int(input("Dime tu edad: "))

if (edad < 18):
  print("¡A estudiar!")
elif edad < 65:
  print("Eres un adulto")
else:
  print("Eres anciano")
```

**Nota importante:** El `match` está disponible a partir de Python `3.10`

```python
numero = int(input("Introduzca un numero entero: "))

match numero:
  case 1:
    print("Uno")
  case 2:
    print("Dos")
  case 3 | 4 | 5:
    print("Es menor o igual que cinco")
  case _:
    print("Numero mayor que cinco")
```

## Funciones

```python
def saluda():
  print("¡Hola!")

saluda()
```

```python
def torre_inclinada(altura):
  for i in range(altura, 0, -1):
    print(" " * (i - 1) + "*")

torre_inclinada(7)
```

### Docstrings

Los *Docstrings* son comentarios esenciales que comienzan y terminan con triples comillas dobles (`"""`). Pueden ocupar varias líneas.

Son el equivalente al *Javadoc* de Java.

```python
def saluda(nombre):
  """Muestra un saludo por pantalla que incluye el usuario"""
  print(f"Buenos días, {nombre}")
```

```python
def torre_cuatro():
  """Pinta una torra con una altura de 4 asteriscos"""
  for i in range (4):
    print("*")

torre_cuatro()
```

```python
def piramide(altura):
  """Pinta una piramide con una altura que se le pasa como argumento."""
  for i in range(altura + 1):    # No se incluye el ultimo valor, aunque sea un rango de 6, se queda en 5, va desde 0 - 5
    print((altura - i) * " "     # Espacios en blanco
          + (i * 2 - 1) * ("*")) # Asteriscos


piramide(5)
```
