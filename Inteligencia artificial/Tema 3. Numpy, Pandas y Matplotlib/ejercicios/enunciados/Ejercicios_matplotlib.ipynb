## Ejercicio 8

Muestra en una misma gráfica, las funciones lineal, cuadrática y cúbica.

Debe aparecer una leyenda indicando a qué corresponde cada línea.

La gráfica debe tener un título.

Las líneas deben ser cada una de un tipo: continua, con pequeñas líneas discontinuas y con puntos.

```python
array = np.linspace(0, 2, 100)

plt.plot(array, array, "-", label="lineal")
plt.plot(array, array**2, "--", label="cuadratica")
plt.plot(array, array**3, ".", label="cubica")

plt.xlabel("x label")
plt.ylabel("y label")

plt.legend()
plt.title("Simple plot")

plt.show()
```

## Ejercicio 9

Representa, en la misma gráfica, una línea a base de círculos, otra hecha con triángulos y otra formada por cuadraditos.

```python
array = np.arange(0., 11., 0.4)

plt.plot(array, array, 'bo', array, array**2, 'r^', array, array**3, 'gs')
```

## Ejercicio 10

Crea una gráfica de color amarillo para la función $y = x² + 1$ en el intervalo de -1 a 1 usando la API orientada a objetos de Matplotlib.

```python
x = np.linspace(1, -1, 100)
y = x ** 2 + 1

fig, ax = plt.subplots()

ax.plot(x, y, color="yellow")

ax.set_title("Graph using Object Oriented API")

plt.show()
```
